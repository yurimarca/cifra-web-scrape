#### Function to return genre of the artist and create a new column to the data frame

```python
for i, cifra in cifras_df.iterrows():
    c_artist_url = cifra["cifra_url"][32:].split("/")[0]
    print(c_artist_url)
    genre = artist_genre_df[artist_genre_df["artist_url"] == c_artist_url]["genre"]
    if genre.empty:
        print("ERROR")
    print(genre.to_string())
    
    print()
```

```python
def add_genre_cifras(genre_df, cifras_df):
    
    new_col = []
    for i, cifra in cifras_df.iterrows():
        c_artist_url = cifra["cifra_url"][32:].split("/")[0]
        genre = genre_df[genre_df["artist_url"] == c_artist_url]["genre"]
        if genre.empty:
            #print("ERROR")
            new_col.append("")
        else:
            new_col.append(genre.to_string().split("    ")[1])
    
    return new_col

new_col = add_genre_cifras(artist_genre_df, cifras_df)
print(new_col)
```

```python
cifras_df["genre"] = new_col
```

### There are some "weird" keys, starting from "+68"
Let's build a function that will return the .txt cifra from the data frame

```python
cifras2_df.columns
```

```python
def return_cifra_txt(cifra_file_loc):
    with open(cifra_file_loc, "r") as file:
        cifra_txt = file.read()
    return cifra_txt
```

```python
sorted_cifra_keys = sorted(cifra_keys_distr,key=lambda x: x[1], reverse=True)
print(sorted_cifra_keys[33])

for i, c in cifras2_df[cifras2_df["cifra_key"]==sorted_cifra_keys[33][0]].iterrows():
    print("-----------------------------------------------------------")
    print("Artist: " + str(c["url_artist"]))
    print("Song: " + str(c["url_song"]))
    print(return_cifra_txt(c["cifra_file_loc"]))
```

```python
for j in range(len(sorted_cifra_keys)-34):
    print("////////////////////////////////////////////////////////////////////////")
    print(sorted_cifra_keys[34+j])

    for i, c in cifras2_df[cifras2_df["cifra_key"]==sorted_cifra_keys[34+j][0]].iterrows():
        print("-----------------------------------------------------------")
        print("Artist: " + str(c["url_artist"]))
        print("Song: " + str(c["url_song"]))
        print(return_cifra_txt(c["cifra_file_loc"]))
```

```python
def is_key_valid(key):
    valid_keys = ["A", "B", "C", "D", "E", "F", "G"]
    valid_keys += ["A#", "C#", "D#", "F#", "G#"]
    valid_keys += ["Ab", "Bb", "Db", "Eb", "Gb"]
    valid_keys += [v+"m" for v in valid_keys]
    
    if key in valid_keys:
        return True
    else:
        return False
valid_keys_only_df = cifras2_df[cifras2_df["key"] == "nao-informada"]
```

```python
valid_keys_only_df = cifras2_df.iloc[[i for i, c in cifras2_df.iterrows() if is_key_valid(c["cifra_key"])]]
```
