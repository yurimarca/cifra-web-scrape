````python
def search_log_file(filename):
    stored_info = []
    with open(filename, 'r') as f:
        for line in f:
            if "Cifra URL #" in line:
                new_info = "https://" + line.split("https://")[-1].split("\n")[0]
            if "Song web title: " in line:
                info = line.split("Song web title: ")[-1].split(", BR=")[0]
                if len(info.split(" - ")) == 2:
                    song_info, artist_info = info.split(" - ")
                else:
                    aux = info.split(" - ")
                    song_info = aux[0]
                    song_info = aux[1]
                
                stored_info.append([new_info, song_info, artist_info])
        return stored_info

df_log_file = pd.DataFrame(search_log_file("cifras_scraper2.log"), 
                           columns=["cifra_url","song_name","artist_name"])
df_log_file.head()
````

#### ```artist_name``` and ```song_name``` information got from the log file

- I could retrieve the information from the log file and store in a data frame.
- Now, we can get ```artist_name``` and ```song_name``` information from another dataframe similarly of how we got information about music genre.
- To do this, we need to generate ```cifra_url``` first and add to ```cifras_df2```.

````python
def get_cifra_url(df):
    url = "https://www.cifras.com.br/cifra/"
    ret_list = [ url + "/".join(file_loc.split(".")[0].split("/")[1:]) for file_loc in df["cifra_file_loc"]]

    return ret_list
    
cifras_df2["cifra_url"] = get_cifra_url(cifras_df2)
cifras_df2.head()
````

````python
def get_artist_name_from_log(df_log_file, cifra_url):
    return df_log_file[df_log_file["cifra_url"] == cifra_url]["artist_name"].to_string().split("    ")[1:]
print(get_artist_name_from_log(df_log_file, "https://www.cifras.com.br/cifra/shirley-carvalhaes/eu-sou-deus"))
````

````python
def get_song_name_from_log(df_log_file, cifra_url):
    return df_log_file[df_log_file["cifra_url"] == cifra_url]["song_name"].to_string().split("    ")[1:]
print(get_song_name_from_log(df_log_file, "https://www.cifras.com.br/cifra/shirley-carvalhaes/eu-sou-deus"))
````

````python
new_column_artist_name = [get_artist_name_from_log(df_log_file, url)  for url in cifras_df2['cifra_url']]
new_column_song_name = [get_song_name_from_log(df_log_file, url)  for url in cifras_df2['cifra_url']]
new_column_song_name[0:4]
````
