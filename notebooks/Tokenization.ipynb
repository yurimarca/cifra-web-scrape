```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('neuralmind/bert-base-portuguese-cased', do_lower_case=False)
```

```python
import pandas as pd

valid_cifras_df = pd.read_csv("valid_cifras.csv", index_col=0)
```

```python
def return_cifra(cifra_df_item):
    cifra_file = cifra_df_item["cifra_file_loc"].item()
    with open(cifra_file, "r") as file:
        cifra = file.read()
    return cifra
```

```python
print(return_cifra(valid_cifras_df.sample()))
```

```python
cifras_text = return_cifra(valid_cifras_df.sample())

# Encode the text using the tokenizer
encoded_text = tokenizer(cifras_text, truncation=True, padding="max_length", max_length=512)

# Extract the input IDs (tokenized text)
input_ids = encoded_text["input_ids"]
```

```python
import re

def extract_chords(cifra_text):
    # Define regular expression pattern to match chords
    chord_pattern = re.compile(r'\b[A-G][#b]?(maj|min|maj7|min7|sus|dim|aug|6|7|9|11|13)?(/[A-G][#b]?)?\b')

    # Find all occurrences of chords in the cifra text
    chords = chord_pattern.findall(cifra_text)

    # Join the chords into a string
    chords_string = ' '.join([''.join(chord) for chord in chords])

    return chords_string

cifra_text = """
Intro F  C/E  Dm7  Bb  
      F/A  Bb  F/C  C4  C  
      F  C/E  Dm7 
      Bb  F/A  Bb7M  C4  

         F          C/E        Dm7  F/C 
Quando a minha esperança acabar 
         Bb         F/A           Gm7  C7(4) 
Quando o Sol pra mim   não brilhar 
          F          Eb6       D7(4)  D7 
Eu só preciso a Jesus  recorrer 
     Gm7     F/A   Bb7M  C4  C 
"""

chords_string = extract_chords(cifra_text)
print(chords_string)
```
